```python
# Importando o numpy e o matplot
import numpy as np
import matplotlib.pyplot as plt

#Criando uma lista de 10000 numeros aleatorios com media 100 e desvio padrao 20
incomes = np.random.normal(100.0, 20.0, 10000)

# criando o histograma e mostrando
plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()  # Desvio padrao
```

```python
incomes.var()  # Variancia
```
