```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def de_mean(x): # função que retorna a diferença entre cada valor e a média dos valores da lista
    xmean = np.mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y): # função que calcula a covariância entre duas listas 
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1) # a função dot calcula o produto escalar entre as duas arrays

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

plt.scatter(pageSpeeds, purchaseAmount) # criando o gráfico de dispersão
plt.show()

covariance(pageSpeeds, purchaseAmount) # calculando a covariância entre as duas listas
```

Olhando o resulta que é bem próximo de 0 e olhando o gráfico é possível ver que não tem muita relação entre os valores

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds # criando uma relação entre as duas listas
plt.scatter(pageSpeeds, purchaseAmount) # criando o gráfico de dispersão
covariance(pageSpeeds, purchaseAmount) # calculando a covariância entre as duas listas
```

Esse é um valor mais distante de 0, então provavelmente tem uma relação, mas não da pra ter certeza de qual

```python
def correlation(x, y): # função que calcula a correlação entre duas listas
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx / stddevy #In real life you'd check for divide by zero hero

correlation(pageSpeeds, purchaseAmount) # calculando a correlação entre as duas listas
```

Usando o numpy agora

```python
np.corrcoef(pageSpeeds, purchaseAmount) # usando o numpy para calcular a correlação entre as duas listas
```

```python
purchaseAmount = 100 - pageSpeeds * 3 # criando uma correlação perfeita
plt.scatter(pageSpeeds, purchaseAmount) # criando o gráfico de dispersão
correlation(pageSpeeds, purchaseAmount) # calculando a correlação entre as duas listas
```
