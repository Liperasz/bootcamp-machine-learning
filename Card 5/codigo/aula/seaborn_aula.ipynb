Using matplot

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv') # importando arquivo csv

gear_counts = df['# Gears'].value_counts() # contando quantos carros tem certa quantidade de marchas

gear_counts.plot(kind='bar') # plotando o gráfico de barras
plt.show()
```

```python
sns.distplot(df['CombMPG']) #histograma que desenha uma linha de densidade por cima
plt.show()
```

```python
df2 = df[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']] # criando um novo dataframe com apenas as colunas de interesse
df2.head()
```

```python
sns.pairplot(df2, hue='Cylinders', height=2.5) # por algum motivo não está mostrando as cilindradas e eu não consegui corrigir isso.
plt.show()
```

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df)
plt.show()
```

```python
sns.jointplot(x="Eng Displ", y="CombMPG", data=df)
plt.show()
```

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)}) # definindo o tamanho da figura
ax=sns.boxplot(x="Mfr Name", y="CombMPG", data=df) # boxplot de consumo por fabricante
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotacionando os nomes dos fabricantes
plt.show()
```

```python
ax = sns.swarmplot(x="Mfr Name", y="CombMPG", data=df) # swarmplot de consumo por fabricante
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotacionando os nomes dos fabricantes
plt.show()
```

```python
ax = sns.countplot(x='Mfr Name', data=df) # countplot de quantidade de carros por fabricante
ax.set_xticklabels(ax.get_xticklabels(), rotation=45) # rotacionando os nomes dos fabricantes
plt.show()
```

```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean') # criando uma tabela pivô
sns.heatmap(df2) # heatmap da tabela pivô
plt.show()
```
