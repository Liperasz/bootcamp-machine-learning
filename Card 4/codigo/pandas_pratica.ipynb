```python
# Criando um dataframe com esses dados
df_sp = pd.DataFrame(data_sp)
df_sp
```

```python
# Definindo que o nome dos jogadores será o index
# Inplace para realizar a alteração de forma definitiva
df_sp.set_index('Nome', inplace=True)
df_sp
```

```python
# Arrumando o time pela maior quantidade de jogos
df_sp.sort_values('Jogos', ascending=False) # Ascending = false serve para colocar em ordem decrescente
```

```python
# Mostrando a parte de cima da tabela mas apenas os números de gols
df_sp['Gols'].head()
```

```python
# Quantos gols o São Paulo fez na temporada?
df_sp['Gols'].sum()
# método sum para somar todos os valores da coluna gols
```

```python
# Quem tem mais participações em gols no são Paulo?
df_sp['G/A'] = df_sp['Gols'] + df_sp['Assistencias']
df_sp
# criando uma nova coluna g/a que tem a soma dos valores de gols e assistencias de cada jogador
```

```python
# Mostrar os jogadores com mais participações em gols na temporada
df_sp.sort_values('G/A', ascending=False).head()
#Sorteando de forma decrescente os jogadores com base nos números de gols
```

```python
# Calcular a quantidade de minutos que cada jogador demora para participar de um gol
#round para arredondar os minutos para números inteiros, utilizando de filtros para não ocorrer divisão por 0
df_sp['Minutos_Participacao'] = round(df_sp[(df_sp['Minutos_Jogados'] > 0) & (df_sp['G/A'] > 0)]['Minutos_Jogados'] / df_sp['G/A'])
df_sp
```

```python
#criando a nova tabela apenas com os minutos para participar de gols de cada jogador.
df_mp = df_sp['Minutos_Participacao'].dropna().sort_values(ascending=True)
df_mp
```

```python
# Excluindo a tabela de minutos para participar do dataframe principal
df_sp.drop(columns='Minutos_Participacao', inplace=True)
df_sp
```

```python
#Criando dataframe dos jogadores do palmereiras
df_sep = pd.DataFrame(data_sep)
df_sep
```

```python
# Definindo o nome dos jogadores do são paulo como coluna denovo
df_sp.reset_index(inplace=True)
df_sp
```

```python
#criando uma nova coluna com os números de camisas dos jogadores
num_sp = [9, 10, 7, 11, 8, 25, 29, 14, 55, 43, 4, 3, 5, 13, 6, 93]
num_sep = [21, 15, 26, 2, 22, 8, 25, 27, 23, 10, 7, 9, 18, 13, 12]
df_sep['Numero'] = num_sep
df_sp['Numero'] = num_sp
```

```python
# Criando uma coluna tipo para cada time
df_sep['Time'] = 'Palmeiras'
df_sp['Time'] = 'Sao Paulo'
```

```python
# Juntando os dados dos dois times em uma tabela só
df = pd.concat([df_sp, df_sep])
df
```

```python
# Definindo um index duplo com número da camisa e time
df.set_index(['Time', 'Numero'], inplace=True)
df
```

```python
# Ajustando novamente a tabela de gols e assistencias
df['G/A'] = df['Gols'] + df['Assistencias']
df
```

```python
#Quantidade de gol que cada clube fez
group = df.groupby('Time')
group['Gols'].sum()
```

```python
# Jogadores acima da média
jam = df[df['G/A'] > df['G/A'].mean()]
jam
```
