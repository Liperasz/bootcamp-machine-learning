```python
# importando os pacotes do sklearn
import numpy as np
from sklearn import model_selection # corrigindo pq o cross_validation foi para model_selection
from sklearn import datasets
from sklearn import svm

# carregando o dataset iris
iris = datasets.load_iris()
```

```python
# dividindo o dataset em treino e teste
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris.data, iris.target, test_size=0.4, random_state=0)

# criando um modelo de previsao SVM para prever as classes do iris usando os dados de treinamento
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)

# avaliando o modelo com os dados de teste
clf.score(X_test, y_test)
```

```python
# fazendo a validação cruzada dando o modelo, o dataset e os valores reais, alem do numero de folds (cv)
scores = model_selection.cross_val_score(clf, iris.data, iris.target, cv=5)

# precisao de cada fold
print(scores)

# media das precisões
print(scores.mean())
```

```python
# testando agora com o kernel poly (projeta os dados em maior dimensão)
clf = svm.SVC(kernel='poly', C=1).fit(X_train, y_train)
scores = model_selection.cross_val_score(clf, iris.data, iris.target, cv=5)
print(scores)
print(scores.mean())
```

```python
# outro modelo com kernel poly
clf = svm.SVC(kernel='poly', C=1).fit(X_train, y_train)

# medida com os dados de teste
clf.score(X_test, y_test)
```

### Activity

The "poly" kernel for SVC actually has another attribute for the number of degrees of the polynomial used, which defaults to 3. For example, svm.SVC(kernel='poly', degree=3, C=1)

```python
clf = svm.SVC(kernel='poly', C=1, degree=5).fit(X_train, y_train)

clf.score(X_test, y_test)

# O score desse foi melhor puramente falando
```
